# Exercício 03

<p style="text-align:justify;">
Crie uma classe que receba o valor de um produto e a porcentagem de desconto, calcule e mostre o valor do desconto e o valor do produto com o desconto. 
</p>

- Observação: o valor do desconto é calculado por meio da fórmula: 
    valor do desconto = valor do produto * percentual de desconto / 100.

```java
import java.util.Scanner;

//public class CalculaDesconto {

    //public static void main(String[] args) {

        // Criando um objeto Scanner para leitura da entrada
        Scanner scanner = new Scanner(System.in);

        // Lendo o valor do produto
        System.out.println("Digite o valor do produto: ");
        double valorProduto = scanner.nextDouble();

        // Lendo a porcentagem de desconto
        System.out.println("\nDigite a porcentagem de desconto (entre 0 e 100): ");
        double porcentagemDesconto = scanner.nextDouble();

        // Variável para armazenar o valor do desconto
        double valorDesconto = 0;

        // Validando a porcentagem de desconto
        if (porcentagemDesconto < 0 || porcentagemDesconto > 100) {
            System.out.println("Porcentagem de desconto inválida. Digite um valor entre 0 e 100.");

            // Encerra o programa se a porcentagem for inválida
            System.exit(0);
        } else {
            // Calculando o valor do desconto
            valorDesconto = valorProduto * porcentagemDesconto / 100;
        }

        // Calculando o valor final do produto
        double valorFinal = valorProduto - valorDesconto;

        // Exibindo os resultados
        System.out.println("\nValor do desconto: R$ " + valorDesconto);
        System.out.println("Valor final do produto: R$ " + valorFinal);
    //}
//}
```

# Exercício 04

<p style="text-align:justify;">
Crie uma classe que receba dois valores inteiros. 
</p>

- O primeiro valor corresponde à quantidade de pontos do líder do campeonato brasileiro de futebol. 

- O segundo valor corresponde à quantidade de pontos do time lanterna. 

<p style="text-align:justify;">
- Considerando que cada vitória vale 3 pontos, calcule o número de vitórias necessárias para que o time lanterna alcance (ou ultrapasse) o líder. 
    
- Por exemplo, supondo que as quantidades de ponto fornecidas sejam 40 e 22, então o número de vitórias apresentada na saída deverá ser 6, pois (40-22) / 3 = 6.
</p>

```java
import java.util.Scanner;

//public class DiferencaVitoriasSemReturn {

    //public static void main(String[] args) {

        // Criando um objeto Scanner para leitura da entrada
        Scanner scanner = new Scanner(System.in);

        // Variáveis para armazenar os pontos
        int pontosLider;
        int pontosLanterna;

        // Lendo a quantidade de pontos do líder
        System.out.println("Digite a quantidade de pontos do líder: ");
        pontosLider = scanner.nextInt();

        // Lendo a quantidade de pontos do lanterna
        System.out.println("\nDigite a quantidade de pontos do lanterna: ");
        pontosLanterna = scanner.nextInt();

        // Validando os pontos digitados
        if (pontosLider < 0 || pontosLanterna < 0) {
            System.out.println("Os valores de pontos devem ser maiores ou iguais a zero.");
        } else {

            // Calculando a diferença de pontos
            int diferencaPontos = pontosLider - pontosLanterna;

            // Calculando o número de vitórias necessárias
            int vitoriasNecessarias = diferencaPontos / 3;

            // Exibindo o resultado
            System.out.println("\nNúmero de vitórias necessárias para o lanterna alcançar o líder: " + vitoriasNecessarias);
        }
    //}
//}
```
